# Part 3. Factor analysis

[Alphalens](https://github.com/quantopian) is a Python Library for performance analysis of predictive (alpha) stock factors
    

"Long-short equity refers to the fact that the strategy is both long and short in the equity market. This is a rather general statement, but has over time grown to mean a specific family of strategies. These strategies rank all stocks in the market using some model. The strategy then goes long (buys) the top $n$ equities of the ranking, and goes short on (sells) the bottom $n$ while maintaining equal dollar volume between the long and short positions. This has the advantage of being statistically robust, as by ranking stocks and entering hundreds or thousands of positions, you are making many bets on your ranking model rather than just a few risky bets. You are also betting purely on the quality of your ranking scheme, as the equal dollar volume long and short positions ensure that the strategy will remain market neutral (immune to market movements)." @ [Lecture on L-S equity](https://github.com/quantopian/research_public/blob/master/notebooks/lectures/Long-Short_Equity/notebook.ipynb)

```python
# setting tickers - sectors dictionary for alphalens
data_sectors = pd.read_csv('./data/sectors.csv').dropna()
sectors = data_sectors.INDUSTRY_SECTOR.unique()
ticker_sector = {sector : data_sectors[data_sectors.INDUSTRY_SECTOR == sector].TICKER.values for sector in sectors}
tickers = ticker_sector.values()
al_tickers_dict = {}
al_sectors_dict = {}
for i,s in enumerate(sectors):
    al_sectors_dict[i] = s
    for t in ticker_sector[s]:
        al_tickers_dict[t] = i
```

Market prices setup

```python
with open('./data/market.pkl','rb') as f:
    data_price_s = pickle.load(f).asfreq('C')

data_price_s.dropna(inplace=True)
data_price = data_price_s.loc[:,(slice(None),slice(None),'close')]#.dropna()
data_price.columns = data_price.columns.droplevel(level=2).droplevel(level=0)

# One sector data
# data_price_s = data_price_s.Financial
# data_price = data_price_s.loc[:,(slice(None),'close')]#.dropna()
# data_price.columns = data_price.columns.droplevel(level=1)
```

Setting up predictive factor

```python
# Ideal factor
# days_ahead = 5
# pred_factor = data_price.pct_change(days_ahead).shift(-days_ahead)#[days_ahead:]

pred_factor = data_price_s.loc[:,(slice(None),slice(None),'rank')]#.rank(axis=1,pct=True,ascending=True)
pred_factor.columns = pred_factor.columns.droplevel(level=2).droplevel(level=0)

# One sector data
# pred_factor = data_price_s.loc[:,(slice(None),'pred')]
# pred_factor.columns = pred_factor.columns.droplevel(level=1)

pred_factor = (pred_factor).stack()
```

```python
#constructing factor data for 1D and 21D ahead
factor_data = alphalens.utils.get_clean_factor_and_forward_returns(pred_factor, 
                                                                   data_price, 
                                                                   quantiles=10,
                                                                   
                                                                   periods=(1,21),
                                                                   bins=None,
                                                                   groupby=al_tickers_dict,
                                                                   groupby_labels=al_sectors_dict)
```

Annual alpha and beta via CAPM model. Alpha is not that great but beta is realy low at .02

```python
q_f = alphalens.utils.quantize_factor(factor_data)
ret = alphalens.performance.factor_returns(factor_data)
alphalens.performance.factor_alpha_beta(factor_data)
```

Mean returns distribution by quantiles 

```python
mret,err = alphalens.performance.mean_return_by_quantile(factor_data)
[alphalens.plotting.plot_quantile_returns_bar(mret[r]) for r in mret.columns]
```

# 3.1 Aftermath

Long-short equity is type of market-neutral strategy that takes long positions in stocks that are expected to appreciate and short positions in stocks that are expected to decline. Thus, heart of this kind of strategy is ranking function, which is designed to minimize market exposure.<br>
Ranking function assign each stock a number based on how they are expected to perform - score. Score can be obtain with various factors, fundamentals, technicals. Strong impact in score can be made with  by machine-learning methods: additional factor usage with sentiment analysis, stacking, boosting, bagging , etc.<br>
